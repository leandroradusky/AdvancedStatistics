```python
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

## Generate datasets for classification and clustering

```python
X, y = make_classification(n_samples=1000, n_features=4,
                            n_informative=2, n_redundant=0,
                            random_state=0, shuffle=False)
```

```python
print (X)
```

```python
print (y)
```

## Random Forest Classifier

```python
clf = RandomForestClassifier(max_depth=2)
clf.fit(X, y)
RandomForestClassifier(max_depth=2, random_state=0)
print(clf.feature_importances_)
```

```python
print(clf.predict([[-1.66853167, -1.29901346,  0.2746472,  -0.60362044]]))
```

```python
print(clf.predict([[1.0415229,  -0.01987143,  0.15216419, -1.9405334]])) 
```

## Decision Tree Classifier

```python
dt_clf = DecisionTreeClassifier()
dt_clf = dt_clf.fit(X, y)
```

```python
fig=plt.figure(figsize=(40, 40))
plot_tree(dt_clf.fit(X,y))
```
