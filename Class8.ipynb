```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
def reduceToTwoDimensionsWithPCAAndPlot(data, data_labels):

    pca = PCA(n_components=2)
    X_r = pca.fit(data).transform(data)

    # Percentage of variance explained for each PCA dimensions
    print('Explained variance: %s' % str(pca.explained_variance_ratio_))

    for label in data_labels:
        plt.scatter(X_r[np.array(data_labels) == label, 0], 
                    X_r[np.array(data_labels) == label, 1], color=label)

    plt.title('PCA')

    plt.figure()
    plt.show()
```

```python
X = norm.rvs(size=1000)
Y = norm.rvs(size=1000)
Z = list(zip(X,Y))
L = ['red']*500+['blue']*500

reduceToTwoDimensionsWithPCAAndPlot(Z, L)
```
